```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 20,15
from myPyLib.Mollweide import moll
```

```python
xao = np.loadtxt('field.xao.out.txt',skiprows=1)
bok = np.loadtxt('field.bok.out.txt',skiprows=1)
```

```python
xra0 = [f[1] for f in xao if f[5] == 0]
xde0 = [f[2] for f in xao if f[5] == 0]
xgl0 = [f[3] for f in xao if f[5] == 0]
xgb0 = [f[4] for f in xao if f[5] == 0]
xra1 = [f[1] for f in xao if f[5] == 1]
xde1 = [f[2] for f in xao if f[5] == 1]
xgl1 = [f[3] for f in xao if f[5] == 1]
xgb1 = [f[4] for f in xao if f[5] == 1]
xra2 = [f[1] for f in xao if f[5] == 2]
xde2 = [f[2] for f in xao if f[5] == 2]
xgl2 = [f[3] for f in xao if f[5] == 2]
xgb2 = [f[4] for f in xao if f[5] == 2]

bra0 = [f[1] for f in bok if f[5] == 0]
bde0 = [f[2] for f in bok if f[5] == 0]
bgl0 = [f[3] for f in bok if f[5] == 0]
bgb0 = [f[4] for f in bok if f[5] == 0]
bra1 = [f[1] for f in bok if f[5] == 1]
bde1 = [f[2] for f in bok if f[5] == 1]
bgl1 = [f[3] for f in bok if f[5] == 1]
bgb1 = [f[4] for f in bok if f[5] == 1]
```

```python
mxe = moll(lat_range=(-25,80))
```

```python
plt.rcParams['figure.figsize'] = 16,12
mxe.grid(lat_lab_lon=0,lon_lab_lat=-22)
mxe.scatter(xra0, xde0, 'bs', label='Abondoned')
mxe.scatter(xra1, xde1, 'rs', label='SDSS')
mxe.scatter(xra2, xde2, 'gs', label='To Do')
plt.legend()
plt.title('XAO Obs Plan in Equatorial System')
plt.savefig('xao.1.eps')
plt.savefig('xao.1.png')
```

```python
mxg = moll(lat_range=(-80,80))
```

```python
plt.rcParams['figure.figsize'] = 16,12
mxg.grid(lat_lab_lon=0,lon_lab_lat=0)
mxg.scatter(xgl0, xgb0, 'bs', label='Abondoned')
mxg.scatter(xgl1, xgb1, 'rs', label='SDSS')
mxg.scatter(xgl2, xgb2, 'gs', label='To Do')
plt.legend()
plt.title('XAO Obs Plan in Galactic System')
plt.savefig('xao.2.eps')
plt.savefig('xao.2.png')
```
