```python
def arrange(days) :
    cover = [0.0] * 12
    for i in range(12) :
        for m in range(i-3, i+3) :
            cover[m%12] += days[i] * 200 / 6.0
    return cover
```

```python
c = arrange([5, 5, 0, 0, 0, 0, 0, 0, 10, 15, 15, 15])
```
